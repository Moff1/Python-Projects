```python
#Question 2
CollegeScoreboard = pd.read_csv("Datasets/MERGED2015_16_PP.csv")
print(CollegeScoreboard)
```

```python
#Question 2a
Public = CollegeScoreboard[CollegeScoreboard["CONTROL"] == 1]
PrivNP = CollegeScoreboard[CollegeScoreboard["CONTROL"] == 2]
PrivFP = CollegeScoreboard[CollegeScoreboard["CONTROL"] == 3]
plt.figure(figsize=(15,15))
 
plt.subplot(311)
plt.hist(Public['GRAD_DEBT_MDN'],rwidth=.5)
plt.title("Grad Debt for Public Colleges")
plt.xlabel("Median Debt")
plt.ylabel("Frequency")

plt.subplot(312)
plt.hist(PrivNP['GRAD_DEBT_MDN'],rwidth=.5)
plt.title("Grad Debt for Private [Not For Profit] Colleges")
plt.xlabel("Median Debt")
plt.ylabel("Frequency")

plt.subplot(313)
plt.hist(PrivFP['GRAD_DEBT_MDN'],rwidth=.5)
plt.title("Grad Debt for Private [For Profit] Colleges")
plt.xlabel("Median Debt")
plt.ylabel("Frequency")

plt.show()

"""
From the plots below I can concluded several trends. For Public Colleges, alot of college graduates have college debt around 10,000 which is actually very surprising to me. 
For Private Non-profit College the most frequent median debt is around 25,000; more than the public which is expected. 
In the case of Private For-profit College, the mode for the median debt is around 10,000 which is unexpected. I would've expected it to be around where the private school level. 
One thing I notice about the range of the debts; it gets larger and larger when moving from Public to Non-profit College to For Profit Colleges.
"""
```

```python
#Question 2b
Public = CollegeScoreboard[CollegeScoreboard["CONTROL"] == 1]
PrivNP = CollegeScoreboard[CollegeScoreboard["CONTROL"] == 2]
PrivFP = CollegeScoreboard[CollegeScoreboard["CONTROL"] == 3]

Public.pivot(columns='ICLEVEL', values='GRAD_DEBT_MDN').plot.hist()
plt.title("Public Schools ICLEVEL and Grad Debt")
plt.xlabel("Median")
plt.ylabel("Frequency")


PrivNP.pivot(columns='ICLEVEL', values='GRAD_DEBT_MDN').plot.hist()
plt.title("Private Not for Profit Schools ICLEVEL and Grad Debt")
plt.xlabel("Median")
plt.ylabel("Frequency")

PrivFP.pivot(columns='ICLEVEL', values='GRAD_DEBT_MDN').plot.hist()
plt.title("Private For Profit Schools ICLEVEL and Grad Debt")
plt.xlabel("Median")
plt.ylabel("Frequency")

plt.show()

"""
From the plots below I can concluded several trends. For Public Colleges, the 4-year colleges have higher median than two-year and less than two year; however 2-year has the highest frequency out of the three categories.  
For Private Non-profit College the most frequent debt comes from the 4-year instituions; private school is not cheap and 4 years adds up fairly quickly. 
In the case of Private For-profit College, the category with the most is 'Less than 2 year schools', but the highest is once again the 4 year schools. 
I think accross the board the 4-year instutions are going to be the highest because you are there for longer and pay more money. 
"""
```
