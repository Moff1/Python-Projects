```python
def calculations(X,Y):
    if len(X) != len(Y):
        print("X and Y are not the same length!")
    else:
        mean_x = np.mean(X)
        mean_y = np.mean(Y)
        n = len(X)
        
        number = 0
        denom = 0
        
        for i in range(n):
            number += (X[i] - mean_x) * (Y[i] - mean_y)
            denom += (X[i] - mean_x) ** 2
            m = number / denom
            c = mean_y - (m * mean_x)
            
            max_x = np.max(X) + 10
            min_x = np.min(X) - 10
            
            x = np.linspace(min_x,max_x, 1000)
            y = c + m * x
            
        dict1 = {"Coeff": m, "Slope":c}
        print(dict1)
    
            
        plt.plot(x,y, color = '#58b970',label = 'Regression Line')
            
        plt.scatter(X,Y, c = '#ef5423', label = 'Scatter Plot')
            
        plt.xlabel("X")
        plt.ylabel("Y")
            
        plt.legend()
        plt.show()
        
        rmse = 0
        
        for i in range(n):
            y_pred = c + m * X[i]
            rmse += (Y[i] - y_pred) ** 2
        rmse = np.sqrt(rmse/n)
        dict2 = {"RMSE": rmse}
        print(dict2)
        
        
        ss_tot = 0
        ss_res = 0
        for i in range(n):
            y_pred = c + m * X[i]
            ss_tot += (Y[i] - mean_y) ** 2
            ss_res += (Y[i] - y_pred) ** 2
        r2 = 1 - (ss_res/ss_tot)
        dict3 = {"R^2": r2}
        print(dict3)
        dict4 = {"Residuals":[]}
        for i in range(n):
            y_pred = c + m * X[i]
            residual = Y[i] - y_pred
            dict4["Residuals"].append(residual)
        print(dict4)
```
